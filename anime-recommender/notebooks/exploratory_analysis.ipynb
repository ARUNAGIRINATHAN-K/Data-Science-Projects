```python
# Cell 1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cell 2
anime = pd.read_csv("/content/anime.csv")
rating = pd.read_csv("/content/rating.csv")

# Cell 3
anime.head()

# Cell 4
rating.head()

# Cell 5
anime['type'].value_counts().plot(kind='bar', title='Anime Type Distribution');

# Cell 6
anime.sort_values(by='members', ascending=False).head(10)[['name', 'members']]

# Cell 7
sns.histplot(anime['rating'].dropna(), bins=20, kde=True)
plt.title('Anime Ratings Distribution')
plt.show()
```
